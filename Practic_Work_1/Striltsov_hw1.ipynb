```python
import pandas as pd
from io import StringIO
import requests
import numpy as np
import matplotlib.pyplot as plt
```

```python

url='https://uk.wikipedia.org/wiki/Народжуваність_в_Україні'
s=requests.get(url).text
data = pd.read_html(StringIO(s))[20]
df = data.head(10)
print(df)
```

```python
num_rows, num_columns = data.shape
print(f"Кількість рядків: {num_rows}, Кількість стовпчиків: {num_columns}")
```

```python
data["1950"] = data["1950"].replace("-", np.nan)
data["1960"] = data["1960"].replace("-", np.nan)
data["1970"] = data["1970"].replace("-", np.nan)
data["1990"] = data["1990"].replace("-", np.nan)
data["1994"] = data["1994"].replace("-", np.nan)
data["1996"] = data["1996"].replace("-", np.nan)
data["2000"] = data["2000"].replace("-", np.nan)
data["2004"] = data["2004"].replace("-", np.nan)
data["2008"] = data["2008"].replace("-", np.nan)
data["2012"] = data["2012"].replace("-", np.nan)
data["2014"] = data["2014"].replace("-", np.nan)
data
```

```python
column_types = data.dtypes
print(column_types)
```

```python
data["2014"] = data["2014"].astype(float)
data.dtypes
```

```python
missing_values_count = data.isnull().sum()
total_values_count = len(data)
# Виведення кількості пропусків у кожному стовпці
for column_name, missing_count in missing_values_count.items():
    print(f"Стовпець '{column_name}': {missing_count} пропусків з {total_values_count}")
```

```python
data = data.drop(data.index[-1])
data
```

```python
data["1950"].fillna(data["1950"].mean(), inplace=True)
data["1960"].fillna(data["1960"].mean(), inplace=True)
data["1970"].fillna(data["1970"].mean(), inplace=True)
data["1990"].fillna(data["1990"].mean(), inplace=True)
data["1994"].fillna(data["1994"].mean(), inplace=True)
data["1996"].fillna(data["1996"].mean(), inplace=True)
data["2000"].fillna(data["2000"].mean(), inplace=True)
data["2004"].fillna(data["2004"].mean(), inplace=True)
data["2008"].fillna(data["2008"].mean(), inplace=True)
data["2012"].fillna(data["2012"].mean(), inplace=True)
data["2014"].fillna(data["2014"].mean(), inplace=True)
data
```

```python
regions_with_higher_birth_rate = data[data['2014'] > data['2014'].mean()]
region_list = regions_with_higher_birth_rate['регіон'].tolist()
print("Список регіонів де народжуваність у 2014 році була вищою за середню по Україні: \n",region_list)
```

```python
# В самій таблиці не було колонки 2010 року, тому було взято 2008
highest_birth_rate_region_2008 = data.loc[data['2008'].idxmax()]['регіон']
highest_birth_rate_region_2012 = data.loc[data['2012'].idxmax()]['регіон']
print("Найвища народжуваність у 2008 році була в регіоні:", highest_birth_rate_region_2008)
print("Найвища народжуваність у 2012 році була в регіоні:", highest_birth_rate_region_2012)
```

```python
plt.figure(figsize=(28, 6))
plt.bar(data['регіон'], data['2014'], color='skyblue')
plt.xlabel('Регіон')
plt.ylabel('Народжуваність у 2014 році')
plt.title('Народжуваність за регіонами у 2014 році')
plt.xticks(rotation=45)
plt.show()
```
