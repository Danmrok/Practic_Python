```python
# Замінюємо колнки великі букви на маленьки, пробіли на підкреслення
df.columns = ['date', 'avila_adobe', 'firehouse_museum', 'chinese_american_museum', 'america_tropical_interpretive_center']
df
```

```python
# За допомогою функції mean дізнаємось середню кількість відвідувачів, а за допомогою round(0) заокргулюємо до цілого числа, а apply(int) переводить з типу float в цілий числовий тип int
average_visitors = df[['avila_adobe', 'firehouse_museum', 'chinese_american_museum', 'america_tropical_interpretive_center']].mean().round(0).apply(int)
average_visitors
```

```python
df['date'] = pd.to_datetime(df['date'])

# Витягуємо рік з дати та створюємо новий стовпець 'Year'
df['Year'] = df['date'].dt.year

# Фільтруємо дані для 2018 року
data_2018 = df[df['Year'] == 2018]

# Знаходимо мінімальну та максимальну кількість відвідувачів для кожного музею за 2018 рік
min_visitors_2018 = data_2018[['avila_adobe', 'firehouse_museum', 'chinese_american_museum', 'america_tropical_interpretive_center']].min()
max_visitors_2018 = data_2018[['avila_adobe', 'firehouse_museum', 'chinese_american_museum', 'america_tropical_interpretive_center']].max()

# Виведіть результат
print("Мінімальна кількість відвідувачів для кожного музею у 2018 році:")
print(min_visitors_2018)

print("\nМаксимальна кількість відвідувачів для кожного музею у 2018 році:")
print(max_visitors_2018)
```

```python
# Витягуєіо місяць з дати
df['Month'] = df['date'].dt.month_name()

# Фільтруємо дані для 2015 року
data_2015 = df[df['Year'] == 2015]

# Згрупуємо дані за місяцем і обчисліть загальну кількість відвідувачів для кожного місяця
monthly_visitors_2015 = data_2015.groupby('Month')[['avila_adobe', 'firehouse_museum', 'chinese_american_museum', 'america_tropical_interpretive_center']].sum()

# Знайдемо місяці з найвищою та найнижчою загальною кількістю відвідувачів
max_month = monthly_visitors_2015.sum(axis=1).idxmax()
min_month = monthly_visitors_2015.sum(axis=1).idxmin()

print("Місяць з найвищою загальною кількістю відвідувачів у 2015 році:", max_month)
print("Місяць з найнижчою загальною кількістю відвідувачів у 2015 році:", min_month)
```

```python
df['date'] = pd.to_datetime(df['date'])

# Витягуємо рік та місяць з дат
df['Year'] = df['date'].dt.year
df['Month'] = df['date'].dt.month

# Фільтруємо дані для 2018 року
data_2018 = df[df['Year'] == 2018]

# Визначаємо літні та зимові місяці
summer_months = [6, 7, 8]  # Червень, Липень, Серпень
winter_months = [12, 1, 2]  # Грудень, Січень, Лютий

# Фільтруємо дані для літніх та зимових місяців
summer_visitors = data_2018[data_2018['Month'].isin(summer_months)]
winter_visitors = data_2018[data_2018['Month'].isin(winter_months)]

# Обчислимо кількість відвідувачів для музею "Avila Adobe" у літні та зимові місяці
summer_visitors_avila = summer_visitors['avila_adobe'].sum()
winter_visitors_avila = winter_visitors['avila_adobe'].sum()


print("Кількість відвідувачів музею 'Avila Adobe' у літні місяці 2018 року:", summer_visitors_avila)
print("Кількість відвідувачів музею 'Avila Adobe' у зимові місяці 2018 року:", winter_visitors_avila)
```

```python
# Фільтруємо дані для 2016 року
data_2016 = df[df['Year'] == 2016]

# Обираємо стовпці з кількістю відвідувачів у музеях
museum_columns = ['avila_adobe', 'firehouse_museum', 'chinese_american_museum', 'america_tropical_interpretive_center']
museum_data = data_2016[museum_columns]

# Обчислення кореляції між кількістю відвідувачів в музеях і датами
correlation_matrix = museum_data.corr()

print("Матриця кореляцій:\n", correlation_matrix)
```

```python
# Будуємо теплокарту кореляції
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Теплокарта кореляції відвідувачів музеїв і дат 2016 року")
plt.show()
```

```python
# Фільтруємо дані для 2017 року
data_2017 = df[df['Year'] == 2017]

# Обираємо необхідні стовпці з кількістю відвідувачів у музеях
museum_columns = ['avila_adobe', 'firehouse_museum', 'chinese_american_museum', 'america_tropical_interpretive_center']

# Будуємо графік ліній для кожного музею
plt.figure(figsize=(12, 6))
for museum in museum_columns:
    plt.plot(data_2017['date'], data_2017[museum], label=museum)

plt.xlabel("Дата")
plt.ylabel("Кількість відвідувачів")
plt.title("Відвідуваність музеїв протягом 2017 року")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Ділимо дані за роками
data_2014 = df[df['Year'] == 2014]
data_2015 = df[df['Year'] == 2015]
data_2016 = df[df['Year'] == 2016]
data_2017 = df[df['Year'] == 2017]

# Колір графіків
colors = ['blue', 'green', 'orange', 'yellow']

# Будуємо графік
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

for i, data in enumerate([data_2014, data_2015, data_2016, data_2017]):
    ax = axes[i // 2, i % 2]
    for j, museum in enumerate(museum_columns):
        data_museum = data.groupby('Month')[museum].sum()
        ax.bar(data_museum.index, data_museum, color=colors[j], alpha=0.7, label=museum)

    ax.set_xlabel("Місяць")
    ax.set_ylabel("Кількість відвідувачів")
    ax.set_title(f"Рік {2014 + i}")
    ax.set_xticks(range(1, 13))
    ax.set_xticklabels([
        "Січ", "Лют", "Бер", "Квіт", "Трав", "Черв",
        "Лип", "Серп", "Вер", "Жовт", "Лист", "Груд"
    ])
    ax.legend()

plt.tight_layout()
plt.show()
```
